```python
import pandas as pd
import re
from collections import defaultdict
import operator
```

```python
def create_dict(file, multiply_by):
    d = defaultdict()
    with open("input_d14.txt") as text:
        for line in text:
            (val, key) = line.strip().split(' => ')
            d[key] = val
    f = defaultdict()
    for i in d.keys():
        a = re.split('(\d+)', d[i])
        a.remove('')
        e = defaultdict()
        j=0
        while j<len(a):
            e[a[j+1].strip().replace(',','')] = int(a[j])
            b=re.split('(\d+)', i)
            f[b[2].strip().replace(',','')] = [e, int(b[1])]
            j = j+2
    f['FUEL'][0] = {k: f['FUEL'][0][k]*multiply_by for k in f['FUEL'][0]}
    f['FUEL'][1] =  f['FUEL'][1]*multiply_by
    return f
```

```python
def ORE_number(f):
    h = {}
    while len(f['FUEL'][0])!=1:
        g = []
        for key in f['FUEL'][0]: 
            value = divmod(f['FUEL'][0][key], f[key][1])
            if key in h:

                if f['FUEL'][0][key]<=h[key]:
                    #g.append((key,f['FUEL'][0][key]))
                    h[key] = h[key]-f['FUEL'][0][key]
                else: 
                    f['FUEL'][0][key] = f['FUEL'][0][key] - h[key]
                    value = divmod(f['FUEL'][0][key], f[key][1])
                    h[key]=0
                    for i in f[key][0]:
                        g.append((i,(value[0]+(1 if value[1]!=0 else 0))*f[key][0][i]))
                        h[key] = max(0,(value[0]+(1 if value[1]!=0 else 0))*f[key][1]-f['FUEL'][0][key])
            else:
                for i in f[key][0]:
                    g.append((i,(value[0]+(1 if value[1]!=0 else 0))*f[key][0][i]))
                    h[key] = max(0,(value[0]+(1 if value[1]!=0 else 0))*f[key][1]-f['FUEL'][0][key])
            #print(g)
        #print(h)
        #print(g)
        totals = defaultdict()
        for key, value in g:
            totals[key] = totals.get(key, 0) + value
        f['FUEL'][0] = totals
    return f['FUEL'][0]['ORE']
```

```python
ore = 0
multiply_by = 1
while ore<1000000000000:
    f = create_dict("input_d14.txt", multiply_by)
    ore = ORE_number(f)
    print(ore, multiply_by)
    multiply_by = multiply_by+50000
    
```
